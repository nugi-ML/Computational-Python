### **Array Creation and Array Elements**

```python
from numpy import linspace

x = linspace(0, 2, 3)

x
```

```python
type(x)
```

```python
type(x[0])
```

```python
sum_elements = x[0] + x[1] + x[2]
sum_elements
```

```python
product_2_elements = x[1] * x[2]
product_2_elements
```

```python
x[0] = 5.0      # overwrite previeous value
x
```

```python
from numpy import zeros

x = zeros(3, int)       # get array with integer zeros

x
```

```python
y = zeros(3)            # get array with floating point zeros

y
```

```python
y[0] = 0.0; y[1] = 1.0; y[2] = 2.0  # overwrite

y
```

```python
len(y)
```

```python
from numpy import array

x = array([0, 1, 2])    # get aray with integers

x
```

```python
x = array([0., 1., 2.])     # get array with real numbers

x
```

```python
from numpy import copy

x = linspace(0, 2, 3)

y = copy(x)

y
```

### **Slicing an Array**

```python
from numpy import linspace

x = linspace(11, 16, 6)

x
```

```python
y = x[1:5]

y
```

```python
y[0] = -1.0

y
```

```python
x
```

### **Two-Dimensional Arrays and Matrix Computations**

```python
import numpy as np

I = np.zeros((3, 3))    # create matrix

I
```

```python
type(I)
```

```python
I[0, 0] = 1.0; I[1, 1] = 1.0; I[2, 2] = 1.0     # Identity matrix

x = np.array([1.0, 2.0, 3.0])       # create vector

y = np.dot(I, x)                    # computes matrix-vector product

y
```

```python
import numpy as np

I = np.eye(3)       # create identity matrix

I
```

```python
type(I)
```

```python
I = np.matrix(I)    # convert to matrix object

type(I)
```

```python
x = np.array([1., 2., 3.,])

x
```

```python
x = np.matrix(x)    # convert to matrix object

x = x.transpose()   # convert to column vector

x
```

```python
y = I * x

y
```
