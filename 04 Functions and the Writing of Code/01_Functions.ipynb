### **Writing Our First Function**

```python
def y(v0, t):
    g = 9.81
    return v0 * t - 0.5 * g * t**2

vo = 5                  # Initial velocity in m/s

time = 0.6              # Just pick one point in time
print(y(vo, time))

time = 0.9              # Pick another point in time
print(y(vo, time))

print(y(10, 1.2))
```

### **A Function with Two Return Values**

```python
def xy(v0x, v0y, t):
    g = 9.81
    return v0x * t, v0y * t - 0.5 * g * t**2

v_init_x = 3
v_init_y = 4
time = 0.5

x, y = xy(v_init_x, v_init_y, time)
print(f"At time {time} seconds, the projectile is at x={x:.2f} m and y={y:.2f} m.")
```

### **A Function with Another Functions as Input Argument**

```python
def f(x):
    return x

def g(x):
    return x**2

def sum_function_values(f, start, stop):
    '''
    Sum up function values for integer argument as
    f(start) + f(start+1) + f(start+2) + ... + f(stop)
    '''
    S = 0
    for i in range(start, stop+1, 1):
        S = S + f(i)
    return S

print(sum_function_values(f, 1, 5))
print(sum_function_values(g, 1, 5))
```

### **Lambda Functions**

```python
'''
def g(x):
    return x**2
'''
# ...which is the same as:
g = lambda x: x**2

print(g(3))
```

```python
print(sum_function_values(lambda x: x, 1, 5))
print(sum_function_values(lambda x: x**2, 1, 5))
```
