```python
import numpy as np

def ask_user(a, b):
    question = f'{a} * {b} = '
    user_input = input(question)
    return int(user_input)

def points(a, b, answer_given):
    correct_answer = a * b
    if answer_given == correct_answer:
        print('Correct!')
        return 1
    else:
        print(f'Incorrect. The correct answer is {correct_answer}.')
        return 0

print('Welcome to the multiplication quiz!')

N = 5
NN = N*N
score = 0
index = list(range(0, NN))
np.random.shuffle(index)
for i in range(0, NN):
    a = index[i]//N + 1
    b = index[i]%N + 1

    user_answer = ask_user(a, b)
    score += points(a, b, user_answer)
    print(f'Your final score is {score} out of {NN}.')

print('Finished!')
```
